```python
import cv2
import matplotlib.pyplot as plt
```

# Draw a Line

`cv2.line(image, start_point, end_point, color, thickness)`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

#specify start and end point of the line
start_point = (135, 25) # (x, y)
end_point = (135, 150) # (x, y)

# specify color and thickness
col = (0, 255, 0)
thick = 3

# draw a line on image
cv2.line(image, start_point, end_point, col, thick)

# display image
plt.imshow(image[:, :, ::-1]), plt.axis('off')
```

# Draw an Arraow

`cv2.arrowedLine(image, start_point, end_point, color, thickness, line_type, shift, tipLength)`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

#specify start and end point of the line
start_point = (135, 25) # (x, y)
end_point = (135, 150) # (x, y)

# specify color and thickness
col = (0, 255, 0)
thick = 3

tip_length = 1

# draw an arrow on image
cv2.arrowedLine(image, start_point, end_point, col, thick, tip_length)

# display image
plt.imshow(image[:, :, ::-1]), plt.axis('off')
```

# Draw an Arraow

`cv2.ellipse(image, centerCoordinates, axesLength, angle, startAngle, endAngle, color [, thickness[, lineType[, shift]]])`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

#specify center coordinates and axes length (ax, ay) of the ellipse
center_coordinates = (120, 100) 
axes_length = (50, 25)

#specify angles
angle = 0
start_angle = 0
end_angle = 360
   
# Red color in BGR
color = (0, 0, 255)
   
# Line thickness
thickness = 5


# draw an ellipse on image
cv2.ellipse(image, center_coordinates, axes_length,
           angle, start_angle, end_angle, color, thickness)

# display image
plt.imshow(image[:, :, ::-1]), plt.axis('off')
```

# Draw a Circle

`cv2.circle(image, center_coordinates, radius, color, thickness)`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

#specify center coordinates and radius of the circle 
center_coordinates = (100, 70)
radius = 30

# Red color in BGR
color = (0, 0, 255)
   
# Line thickness
thickness = 2


# draw an ellipse on image
cv2.circle(image, center_coordinates, radius, color, thickness)

# draw an ellipse on image
cv2.circle(image, (170, 70), radius = 30, color = (0, 255, 0), thickness = -1) # '-1' thickness fills up the circle

# display image
plt.imshow(image[:, :, ::-1])#, plt.axis('off')
```

# Draw a Rectangle

`cv2.relectang(image, start_point, end_point, color, thickness)`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

#specify top-left and bottom-right coordinates of the rectangle
tl = (50, 40) # (xmin, ymin)
br = (200, 100) # (xmax, ymax)

# Red color in BGR
color = (0, 0, 255)
   
# Line thickness
thickness = 2

# draw a rectangle on image
cv2.rectangle(image, tl, br, color, thickness)

# display image
plt.imshow(image[:, :, ::-1]), plt.axis('off')
```

# Draw a Text on Image

`cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])`

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

# text to be inserted
text = "Two cute Puppies"

# origin coordinates to put the text
org = (50, 150)

# font of the text
font = cv2.FONT_HERSHEY_SIMPLEX

# font scale
font_scale = 0.6

# yellow color in BGR
color = (0, 255, 255)
   
# Line thickness
thickness = 2

cv2.putText(image, text, org, font, font_scale, color, thickness)

# display image
plt.imshow(image[:, :, ::-1]), plt.axis('off')
```
