```python
# import all necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Reading Image and displaying using OpenCV

cv2.imread(filename, flag)  
**filename**: path to the image  
**flag**:   
- cv2.IMREAD_UNCHANGED or -1 
- cv2.IMREAD_GRAYSCALE  or 0
- cv2.IMREAD_COLOR  or 1 (default)  

more about flags [link](https://docs.opencv.org/master/d8/d6a/group__imgcodecs__flags.html#ga61d9b0126a3e57d9277ac48327799c80)

cv2 reads the image in BGR format and normal images are in RGB format


```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg")

# print shape of the image
print(image.shape) # returns (height, width, channels)

# show the image in cv2 window
cv2.imshow(winname = "Dogs", mat = image)

# waitKey() waits for a key press to close the window and 0 specifies indefinite loop
cv2.waitKey(0)

# cv2.destroyAllWindows() simply destroys all the windows we created.
cv2.destroyAllWindows()

# save the image
cv2.imwrite(filename = "data/2dogs.jpeg", image = image)
```

## Displaying the image using matplotlib

```python
# Load the image
image = cv2.imread(filename = "data/2dogs.jpeg") # BGR image

# show image
plt.imshow(image) # displays BGR image
```

```python
# convert BGR image to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(rgb_image) # displays RGB image
```

```python
# altername way to convert BGR to RGB
alt_rgb_image = image[:, :, ::-1]

# show image
plt.imshow(alt_rgb_image) # displays RGB image
```

## matplotlib tricks

```python
plt.figure(figsize = (10, 5)) # size of the matplotlib plot (width, height)

# for showing multiple plots in a single figure
# plt.subplot(row_count, column_count, plot_index)

plt.subplot(1, 2, 1), plt.imshow(image), plt.axis('off'), plt.title("BGR Image")
plt.subplot(1, 2, 2), plt.imshow(rgb_image), plt.axis('off'), plt.title("RGB Image")
```
