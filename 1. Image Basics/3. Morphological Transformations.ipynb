```python
# import all necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Morphological Transformations

### Erosion

```python
# Load the image
image = cv2.imread("data/1.jpg", cv2.IMREAD_GRAYSCALE)


# define a kernel, larger the kernel size, more will be the erosion
kernel = np.ones((5,5),np.uint8)

# erode the image
erosion = cv2.erode(image,kernel,iterations = 1)

# display both rotated images
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap = 'gray'), plt.axis('off'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(erosion, cmap = 'gray'), plt.axis('off'), plt.title("erosion")
```

### Dilation

```python
# Load the image
image = cv2.imread("data/1.jpg", cv2.IMREAD_GRAYSCALE)


# define a kernel, larger the kernel size, more will be the dilation
kernel = np.ones((5,5),np.uint8)

# erode the image
dilated = cv2.dilate(image,kernel,iterations = 1)

# display both rotated images
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap = 'gray'), plt.axis('off'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(dilated, cmap = 'gray'), plt.axis('off'), plt.title("dilated")
```

### opening

**Opening** is erosion followed by dilation. Mostly used to remove noise

```python
# Load the image
image = cv2.imread("data/noisy1.jpg", cv2.IMREAD_GRAYSCALE)


# define a kernel, larger the kernel size, more will be the opening
kernel = np.ones((5,5),np.uint8)

# open the image
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# display both rotated images
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap = 'gray'), plt.axis('off'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(opening, cmap = 'gray'), plt.axis('off'), plt.title("Opening")

```

### Closing

Closing is reverse of openning. It is dilation followed by erosion

```python
# Load the image
image = cv2.imread("data/noisy11.jpg", cv2.IMREAD_GRAYSCALE)


# define a kernel, larger the kernel size, more will be the closing
kernel = np.ones((5,5),np.uint8)

# open the image
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# display both rotated images
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap = 'gray'), plt.axis('off'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(closing, cmap = 'gray'), plt.axis('off'), plt.title("Closing")

```

### Morphological Gradient

It is the difference between dilation and erosion of an image.

```python
# Load the image
image = cv2.imread("data/1.jpg", cv2.IMREAD_GRAYSCALE)


# define a kernel, larger the kernel size, more will be the gradient
kernel = np.ones((5,5),np.uint8)

# open the image
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

# display both rotated images
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap = 'gray'), plt.axis('off'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(gradient, cmap = 'gray'), plt.axis('off'), plt.title("Gradient")

```
