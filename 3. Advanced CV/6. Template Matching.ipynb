```python
# import all necessary libraries
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A simple function to display the image using matplotlib
def show(img, cmap = None, figsize = (7, 7)):
    plt.figure(figsize = figsize)
    if cmap is not None:
        plt.imshow(img, cmap = cmap), plt.axis('off')
    else:
        plt.imshow(img), plt.axis('off')
```

# Template Matching

- Template Matching is a method for searching and finding the location of a template image in a larger image
- OpenCV comes with a function `cv.matchTemplate()` for this purpose. It simply slides the template image over the input image (as in 2D convolution) and compares the template and patch of input image under the template image. Several comparison methods are implemented in OpenCV. (You can check docs for more details). 
- It returns a grayscale image, where each pixel denotes how much does the neighbourhood of that pixel match with template.

```python
# Load Image
image = cv2.imread("data/objects.png")
show(image[:, :, ::-1])
```

```python
# Load template to match
ball = cv2.imread("data/ball_template.png")
show(ball[:, :, ::-1], figsize = (2, 2))
```

## Template Matching Methods

1. TM_SQDIFF
2. TM_SQDIFF_NORMED
3. TM_CCORR
4. TM_CCORR_NORMED
5. TM_CCOEFF
6. TM_CCOEFF_NORMED

**Note: To read more about these methods visit [link](https://docs.opencv.org/3.4/de/da9/tutorial_template_matching.html)**

```python
# Define all the methods in a list
methods = ['cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED', 'cv2.TM_CCORR', 
           'cv2.TM_CCORR_NORMED', 'cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED']
```

### Lets try out first method of match template 

```python
# use cv2's in-built matchTemplate function, 
# it return heatmap where the template is matching with the full image
heatmap = cv2.matchTemplate(image, ball, cv2.TM_CCOEFF)

# use minMaxLoc function to get the location of brightspot on the heatmap
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(heatmap)
temp_height, temp_width, _ = ball.shape

# define top-left and bottom-right corners to draw a rectangle
tl = max_loc
br = (tl[0]+temp_width, tl[1]+temp_height)

# make a copy of original image so to draw a rectangle
draw = image.copy()
cv2.rectangle(draw, tl, br, (0, 0, 255), 3)

# display heatmap as well as original image
plt.figure(figsize = (15, 5))
plt.subplot(1, 2, 1), plt.imshow(heatmap), plt.axis('off'), plt.title("Heatmap");
plt.subplot(1, 2, 2), plt.imshow(draw[:, :, ::-1]), plt.axis('off'), plt.title("Matched template");
```

### compare all the methods

```python
for m in methods:
    
    method = eval(m)
    heatmap = cv2.matchTemplate(image, ball, method)
    
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(heatmap)
    temp_height, temp_width, _ = ball.shape
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        tl = min_loc # top-left
    else:
        tl = max_loc
        
    br = (tl[0]+temp_width, tl[1]+temp_height) # bottom-right
    
    draw = image.copy()
    cv2.rectangle(draw, tl, br, (0, 0, 255), 3)

    # display heatmap as well as original image
    plt.figure(figsize = (15, 5))
    plt.subplot(1, 2, 1), plt.imshow(heatmap), plt.axis('off'), plt.title("Heatmap");
    plt.subplot(1, 2, 2), plt.imshow(draw[:, :, ::-1]), plt.axis('off'), plt.title("Matched template");
    plt.suptitle(str(m))
```
