```python
# import all necessary libraries
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

# Contour Detection example

- Contour detection is to detect border of the objects in the image. Contour detection can be used in background substraction, object detection, segmentation etc. 
- to learn more about the contours, visit [link](https://opencv24-python-tutorials.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_contours/py_contours_begin/py_contours_begin.html#contours-getting-started)

### Steps for Contour Detection

1. Load the original image and convert it to grayscale image
2. Applying thresholding (**cv2.threshold**) to grayscale image such that all the object will be in foreground
3. Once thresholding is done, used **closing** morphological operation to remove any background pathches from the foreground object
4. After closing, use `cv2.findContours(thresh, mode, method)` to find the countours from closing image. 
5. Next step is to draw the contours on original image. For that use `cv2.drawContours(draw, contours, contourIdx, color, thickness)` to draw the contours on the image

```python
# A simple function to display the image using matplotlib
def show(img, cmap = None, figsize = (7, 7)):
    plt.figure(figsize = figsize)
    if cmap is not None:
        plt.imshow(img, cmap = cmap), plt.axis('off')
    else:
        plt.imshow(img), plt.axis('off')
```

```python
# Load and display original Image
image = cv2.imread("data/coins.jpg")
show(image[:, :, ::-1])
```

```python
# Convert the image to grayscale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show(gray, cmap = 'gray')
```

```python
# set threshold value and max value. Can only be decided by trial and error method
thresh = 150
maxvalue = 255

# Threshold grayscale image and display 
_, thresh = cv2.threshold(gray, thresh, maxvalue, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
show(thresh, cmap = 'gray')
```

```python
# To remove black pixels from while foreground use closing morphological operation 
# define a kernel. Kernel size and iterations can be decided by trial-and-error.
kernel = np.ones((15, 15), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations = 5)
show(closing, cmap = 'gray')
```

```python
# After getting clean thresholded image, find contours.
cnts, _ =cv2.findContours(thresh, mode = cv2.RETR_EXTERNAL, method = cv2.CHAIN_APPROX_SIMPLE)
draw = image.copy()
cv2.drawContours(draw, contours = cnts, contourIdx = -1, color = (0, 0, 255), thickness = 10)
show(draw[:, :, ::-1])
```

# More Contour Functionalities

```python
# Load the image and display
cat = cv2.imread("data/cat_logo.png")
show(cat, figsize = (4, 4))
```

```python
# Follow all above mentioned steps to find and draw contours
gray = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
cnts, _ =cv2.findContours(thresh, mode = cv2.RETR_EXTERNAL, method = cv2.CHAIN_APPROX_NONE)
draw = cat.copy()
cv2.drawContours(draw, contours = cnts, contourIdx = -1, color = (0, 0, 255), thickness = 2)
show(draw[:, :, ::-1], figsize = (4, 4))
```

## Finding Contour Area

- Contour area is given by the function `cv2.contourArea(cnt)`

```python
# after finding contour, find area of contour
cat_contour = cnts[0]
contour_area = cv2.contourArea(cat_contour)
print(f"Contour area is {contour_area} sq.px")
```

# Finding Contour perimeter

- Contour perimeter can be found out using `cv2.arcLength()` function. Second argument specify whether shape is a closed contour (if passed True), or just a curve

```python
cat_contour = cnts[0]
contour_perimeter = cv2.arcLength(cat_contour,True)
print(f"Contour perimeter is {contour_perimeter} px")
```

## Bounding Rectangle

```python
cat_contour = cnts[0]
x,y,w,h = cv2.boundingRect(cat_contour)

draw = cat.copy()
draw = cv2.rectangle(draw,(x,y),(x+w,y+h),(0,0,255),2)
show(draw[:, :, ::-1], figsize = (4, 4))
```

# Rotated bounding Rectangle

```python
cat_contour = cnts[0]

rect = cv2.minAreaRect(cat_contour)
box = cv2.boxPoints(rect)
box = np.int0(box)

draw = cat.copy()
draw = cv2.drawContours(draw,[box],0,(0,0,255),2)
show(draw[:, :, ::-1], figsize = (4, 4))
```

### Contour Approximation

- It approximates a contour shape to another shape with less number of vertices depending upon the precision we specify
- second argument is called epsilon, which is maximum distance from contour to approximated contour. It is an accuracy parameter. A wise selection of epsilon is needed to get the correct output

```python
cat_contour = cnts[0]
epsilon = 0.05*cv2.arcLength(cat_contour,True)
approx_cnt = cv2.approxPolyDP(cat_contour,epsilon,True)

draw = cat.copy()
cv2.drawContours(draw, contours = approx_cnt, contourIdx = -1, color = (0, 0, 255), thickness = 4)
show(draw[:, :, ::-1], figsize = (4, 4))
```

## Minimum Enclosing Circle

```python
cat_contour = cnts[0]

(x,y),radius = cv2.minEnclosingCircle(cat_contour)
center = (int(x),int(y))
radius = int(radius)

draw = cat.copy()
cv2.circle(draw,center,radius,(0,0,255),2)
show(draw[:, :, ::-1], figsize = (4, 4))
```
