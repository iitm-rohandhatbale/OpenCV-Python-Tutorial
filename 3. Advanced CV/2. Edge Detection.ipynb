```python
# import all necessary libraries
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A simple function to display the image using matplotlib
def show(img, cmap = None, figsize = (7, 7)):
    plt.figure(figsize = figsize)
    if cmap is not None:
        plt.imshow(img, cmap = cmap), plt.axis('off')
    else:
        plt.imshow(img), plt.axis('off')
```

# Canny Edge Detection

Steps involved internally in **Canny Edge Detection** are as follows:
  1. Apply Gaussian filter to smooth the image and remove noise
  2. Find the intensity gradient of the image
  3. Apply non max suppression to get rid of spurious response to edge detection
  4. Apply double threshold to determine potential edges
  5. Track edge y histerisis: Finaloze the detection of edges by suppressing all other edges that are weak and not connected to strong edges
 
**Note: For High resolution images, it is good idea to apply custom blur before uing canny edge detection method**

```python
# Load and display image
image = cv2.imread("data/objects.png")
show(image[:, :, ::-1])
```

```python
# lower threshold value and upper threshold value required to find the edges using cv2.Canny. 
# These values are determined by trial and error method but there is also a thumb rule to find. 

# Lower threshold is max(0, 70% of median pixel value)
lower_thresh = int(max(0, 0.7*np.median(image))
)
# upeer threshold is min(255, 130% of median pixel value)
upper_thresh = int(min(255, 1.3*np.median(image)))

print(f"Lower threshold is {lower_thresh} && upper threshold is {upper_thresh}")
```

```python
# Use canny edge detection with above lower and upper threshold values and display edges
edges = cv2.Canny(image, threshold1 = lower_thresh, threshold2 = upper_thresh)
show(edges, cmap= 'gray')
```

```python
# Load and display nature image
nature_image = cv2.imread("data/nature.png")

print(f"shape: {nature_image.shape}")
show(nature_image[:, :, ::-1])
```

```python
# Try above mentioned thumb-rule to determine upper and lower value of edges

# Lower threshold is max(0, 70% of median pixel value)
lower_thresh = int(max(0, 0.7*np.median(nature_image))
)
# upeer threshold is min(255, 130% of median pixel value)
upper_thresh = int(min(255, 1.3*np.median(nature_image)))

print(f"Lower threshold is {lower_thresh} && upper threshold is {upper_thresh}")
```

```python
# Use canny edge detection with above lower and upper threshold values and display edges
edges = cv2.Canny(nature_image, threshold1 = lower_thresh, threshold2 = upper_thresh)
show(edges, cmap= 'gray')
```

<div class="alert alert-block alert-info">
<b>Observation:</b> As can be seen, there are lot of redundant edges in the image. 
    To remove thse redundant edges, blur the image and then use Canny edge detection
</div>

```python
# Use gaussian blur to smooth image and remove noise, apply cany edge detection and display image
smooth_nature_image = cv2.GaussianBlur(nature_image, (7, 7), 0)
edges = cv2.Canny(smooth_nature_image, threshold1 = lower_thresh, threshold2 = upper_thresh)
show(edges, cmap= 'gray')
```
